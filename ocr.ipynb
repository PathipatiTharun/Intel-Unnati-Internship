```python
# OCR/ML Model Evaluation Metrics Demo

#This notebook demonstrates how to evaluate an OCR/ML model for serial number or label extraction, including accuracy, precision, recall, F1 score, inference time, and confidence.
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
import matplotlib.pyplot as plt
```

```python
import time
```

```python
# Simulate ground truth serial numbers
y_true = ['ABC1234', 'XYZ5678', 'LMN9012', 'DEF3456', 'PQR7890']
```

```python
# Simulate model predictions
y_pred = ['ABC1234', 'XYZ5678', 'LMN9012', 'DEF3450', 'PQR7890']
```

```python
# Simulate confidence scores
confidences = [0.98, 0.95, 0.92, 0.60, 0.99]
```

```python
# Simulate inference times (seconds)
inference_times = [0.8, 0.7, 0.9, 1.1, 0.85]
```

```python
accuracy = accuracy_score(y_true, y_pred)
```

```python
precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
```

```python
recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
```

```python
f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
```

```python
avg_confidence = np.mean(confidences)
```

```python
avg_inference_time = np.mean(inference_times)
```

```python
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
print(f'Average Confidence: {avg_confidence:.2f}')
print(f'Average Inference Time: {avg_inference_time:.2f} s')
```

```python
labels = list(set(y_true + y_pred))
```

```python
cm = confusion_matrix(y_true, y_pred, labels=labels)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(cm, cmap='Blues')
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.colorbar(im)
plt.show()
```

```python
print(classification_report(y_true, y_pred, zero_division=0))
```
